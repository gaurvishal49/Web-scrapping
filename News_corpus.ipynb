```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
```

```python
news_links = []
Id = []
count = 0
for i in range(100):
    page = requests.get('https://www.thehindu.com/news/international/?page='+str(i))
    soup = BeautifulSoup(page.content, 'html.parser')
    headlines = soup.find_all('a', class_='Other-StoryCard-heading')
    for link in headlines:
        Id.append(count)
        news_links.append(link.get('href'))
        count = count+1
```

```python
web_add = []
title = []
Author = []
Date = []
Content = []
for link in news_links:
    page = requests.get(link)
    soup = BeautifulSoup(page.content, 'html.parser')
    try:
        head = soup.find_all('h1', class_='title')[0].get_text().strip()
        name = soup.find_all(class_='auth-nm no-lnk')[0].get_text().strip()
        time = soup.find_all('span', class_='blue-color ksl-time-stamp')[1].get_text().strip()
        cont = soup.find_all('div', class_='paywall')[0].get_text().strip()
        title.append(head)
        Author.append(name)
        Date.append(time)
        Content.append(cont)
        web_add.append(link)
    except:
        pass
```

```python
lists = [title, Author, Date, web_add, Content]
Dat = np.array(lists).T
Data = pd.DataFrame(data=Dat, columns=['Title', 'Author', 'Date', 'News_links', 'Content'])
```
